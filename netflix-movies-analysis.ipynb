```python
df.drop(['show_id'],axis=1,inplace=True)
df;
```

```python
df['type'].value_counts()
```

```python
df.replace({'type':{'Movie':0,'TV Show':1}},inplace =True)
df['title'].value_counts()
```

```python
df = df.drop(columns=['title','director', 'cast'],axis=1)
df
```

```python
df['country'].isnull().sum()
```

```python
df['country'].isnull().value_counts()
```

```python
df['country'] = df['country'].replace(np.nan, 'unknown')
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['date_added','rating','duration'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df['date_added'] = df['date_added'].apply(pd.to_datetime)
df['date_added'].dtype
```

```python
df['loading_year'] = df['date_added'].dt.year
df['loading_year']
```

```python
df['loading_month'] = df['date_added'].dt.month_name()
df['loading_month']
```

```python
df.drop('date_added', axis=1, inplace=True)
```

```python
df
```

```python
df['loading_month'].value_counts()
```

```python
df.replace({'loading_month':{'January':1,'February':2,'March':2,'April':4,'May':5,'June':6,'July':7,'August':8,'September':9,
                             'October':10,'November':11,'December':12}},inplace=True)
df['loading_month']
```
